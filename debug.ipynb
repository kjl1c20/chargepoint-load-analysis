```python
import pandas as pd
from src.utils import get_latest_snapshot_id, load_data, PROCESSED_DATA_DIR

snapshot_id = get_latest_snapshot_id()
DATA_PATH = f"./data/raw/{snapshot_id}.parquet"
df = pd.read_parquet(DATA_PATH)

df["connector_id"]
```

```python
non_convertible = {}

for col in df.select_dtypes(include="object").columns:
    converted = pd.to_numeric(df[col], errors="coerce")
    failed_mask = converted.isna() & df[col].notna()
    if failed_mask.any():
        non_convertible[col] = df[col][failed_mask].unique().tolist()

non_convertible
```
